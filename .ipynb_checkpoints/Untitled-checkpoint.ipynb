```python
select_cols = ['survived', 'name', 'pclass', 'sex', 'age', 'embarked']
selected_df_data = df_data[select_cols]
```

```python
# 填补缺失的 age 值，填充为平均值
age_mean_value = selected_df_data['age'].mean()
selected_df_data['age'] = selected_df_data['age'].fillna(age_mean_value)
```
