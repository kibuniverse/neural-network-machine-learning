```python
%matplotlib notebook

import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
tf.disable_eager_execution()
### 1. è¯»å–æ•°æ®æ–‡ä»¶

df = pd.read_csv("data/boston.csv", header=0)

# æ˜¾ç¤ºæ•°æ®æ‘˜è¦ä¿¡æ¯
# print(df.describe())

# æ˜¾ç¤ºæ‰€æœ‰æ•°æ®
# print(df)

df = df.values # èŽ·å–å€¼ï¼ŒäºŒç»´æ•°ç»„
# print(df)
df = np.array(df) # è½¬ä¸ºnumpyæ•°ç»„

# å½’ä¸€åŒ–æ•°æ®ï¼Œæ‰€æœ‰æ•°æ®èŒƒå›´éƒ½æžåˆ°[0,1]
# ç”±äºŽæ¯ä¸ªç‰¹å¾å¯¹æ ‡ç­¾çš„å½±å“è´¡çŒ®ä¸åŒï¼Œç”šè‡³å·®å¼‚å·¨å¤§ï¼Œå¯¼è‡´å¯¹åº”çš„æƒé‡å°±å·®è·å·¨å¤§
# å¦‚æžœæŸä¸ªç‰¹å¾å¯¹æ ‡ç­¾çš„å½±å“è´¡çŒ®å¾ˆå¤§ï¼Œé‚£ä¹ˆå¯¹åº”çš„æƒé‡å°±å¾ˆå¤§ï¼Œå°±æœ‰å¯èƒ½æƒé‡è¶Šç•Œã€‚ã€‚
for i in range(12):
    df[:, i] = (df[:, i] - df[i:, i].min()) / (df[:, i].max() - df[i:, i].min())

# å‰12åˆ—ç‰¹å¾æ•°æ®(0-11)
x_data = df[:, :12]
# æœ€åŽä¸€åˆ—æ ‡ç­¾æ•°æ®(12)
y_data = df[:, 12]

# print(x_data, x_data.shape)
# print(y_data, y_data.shape)



### 2. å»ºç«‹æ¨¡åž‹

x = tf.placeholder(tf.float32, [None, 12], name="X") # 12åˆ—ç‰¹å¾æ•°æ®ï¼Œè¡Œæ•°ä»»æ„
y = tf.placeholder(tf.float32, [None, 1], name="Y") # 1åˆ—æ ‡ç­¾æ•°æ®ï¼Œè¡Œæ•°ä»»æ„

# å®šä¹‰ä¸€ä¸ªå‘½åç©ºé—´ï¼ŒModelä¸‹çš„æ‰€æœ‰è¯­å¥æˆä¸ºä¸€ä¸ªæ•´ä½“ï¼Œå¯¹åº”åˆ°è®¡ç®—å›¾ä¸­çš„ä¸€ä¸ªå­å›¾
with tf.name_scope("Model"):
    # wåˆå§‹åŒ–ä¸ºshape=(12,1)çš„åˆ—å‘é‡
    w = tf.Variable(tf.random_normal([12,1], stddev=0.01), name="W")
    b = tf.Variable(1.0, name="b")
    
    # å®šä¹‰æ¨¡åž‹ï¼ˆå¤šå…ƒçº¿æ€§å›žå½’ï¼‰
    def model(x, w, b):
        return tf.matmul(x, w) + b
    
    # é¢„æµ‹å€¼
    pred = model(x, w, b)


    
### 3. è®­ç»ƒæ¨¡åž‹

# è¿­ä»£ä¼¦æ¬¡
train_epochs = 50
# å­¦ä¹ çŽ‡
learning_rate = 0.01
# å®šä¹‰æŸå¤±å‡½æ•°
with tf.name_scope("LossFunction"):
    loss_function = tf.reduce_mean(tf.pow(y-pred, 2))
# åˆ›å»ºä¼˜åŒ–å™¨
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss_function)

sess = tf.Session()

init = tf.global_variables_initializer()

sess.run(init)

# éšç€è®­ç»ƒè½®æ•°å˜åŒ–è¿›è¡ŒæŸå¤±å¯è§†åŒ–
loss_list = [] 

for epoch in range(train_epochs):
    loss_sum = 0 # è®¡ç®—æ¯è½®æŸå¤±
    for xs, ys in zip(x_data, y_data):
        xs = xs.reshape(1, 12) # æŠŠä¸€ç»´å‘é‡æžæˆäºŒç»´çš„ï¼Œè¦å¡«å……åˆ°x
        ys = ys.reshape(1, 1)
        _, loss = sess.run([optimizer, loss_function], feed_dict={x:xs, y:ys})
        
        loss_sum += loss

    wtmp = w.eval(session=sess) # æ¯è½®è®­ç»ƒä¹‹åŽçš„w
    btmp = b.eval(session=sess)
    loss_average = loss_sum / len(y_data) # æ¯è½®å¹³å‡æŸå¤±
    loss_list.append(loss_average)

#     æ‰“å°æ¯ä¸€è½®è®­ç»ƒå¾—åˆ°çš„ä¿¡æ¯
#     print("epoch=", epoch+1, "\n", "loss_average=", loss_average, "\n", \
#           "w=", wtmp, "\n", "b=", btmp)
    
    # æ¯è½®è®­ç»ƒä¹‹åŽï¼Œæ‰“ä¹±æ•°æ®é¡ºåºï¼Œä¸ºä¸‹è½®è®­ç»ƒåšå‡†å¤‡ã€‚
    # å°±åƒè¯¾ç¨‹ä¸­å´è€å¸ˆæ‰€è®²ï¼Œè®©ä½ å­¦ä¹ è®¤è¯†å›¾ç‰‡ä¸Šçš„ç”²éª¨æ–‡ï¼Œå¦‚æžœä½ è®°ä½äº†ç¬¬å‡ å¼ å›¾ç‰‡ä¸Šçš„
    # ç”²éª¨æ–‡æ˜¯å•¥ï¼Œé‚£å°±ä¸éœ€è¦çœŸçš„è®¤è¯†ç”²éª¨æ–‡äº†ï¼Œåªéœ€è¦çŸ¥é“å›¾ç‰‡æ˜¯ç¬¬å‡ å¼ ï¼Œå°±çŸ¥é“å¯¹åº”çš„
    # ç”²éª¨æ–‡æ˜¯å•¥äº†ã€‚ã€‚ç„¶è€Œè¿™å¹¶æ²¡æœ‰çœŸæ­£è®¤è¯†ç”²éª¨æ–‡ã€‚ç±»æ¯”ï¼Œæ¯è½®è®­ç»ƒç»“æŸï¼Œæ‰“ä¹±æ•°æ®é¡ºåºã€‚
    x_data, y_data = shuffle(x_data, y_data) # æ‰“ä¹±æ•°æ®
    
# æ‰“å°æŸå¤±å¯è§†åŒ–
plt.plot(loss_list)


### 4. é¢„æµ‹

n = np.random.randint(506)
x_test = x_data[n]
x_test = x_test.reshape(1, 12)
predict = sess.run(pred, feed_dict={x:x_test})
print("n=", n)
print("æ ‡ç­¾å€¼=%f" % y_data[n])
print("é¢„æµ‹å€¼:)=%f" % predict)
```
